## Criando as funcoes

```python
def PC(matriz_u):
    U_quad = matriz_u**2
    soma = U_quad.sum(axis=0)
    soma_final = soma.sum(axis = 0)
    final= (1/len(matriz_u))*soma_final
    return final
```

```python
def pc(matriz_u):
    soma=0
    for i in range(len(matriz_u)):
        ui=u[i]
        soma1=0
        for k in range(12):
            uik=ui[k]
            uik_2 =uik**2
            soma1+= uik_2
        soma+=soma1
    final = (1/len(matriz_u))*soma
    return final
```

```python
def MPC(PC):
    conta = 1- (12/11)*(1-PC)
    return conta
```

```python
def mpc(pc):
    conta = 1- (12/11)*(1-pc)
    return conta
```

```python
def PE(matriz_u):
    soma= matriz_u.sum(axis=0)
    soma_again = soma.sum( axis=0)
    final = (1/len(matriz_u))*soma_again
    return final
    
```

```python
def pe(matriz_u):
    soma=0
    for i in range(len(matriz_u)):
        ui=u[i]
        soma1=0
        for k in range(12):
            uik=ui[k]
            uik_log = uik*np.log(uik)
            soma1+= uik_log
        soma+=soma1
    final = -(1/len(matriz_u))*soma
    return final
```

### fazendo a conta

```python
def fmeasure(matriz_confusao):
    l,_ = matriz_confusao.shape
    s = matriz_confusao.sum()
    s_true = matriz_confusao.sum(axis = 1)
    s_predict = matriz_confusao.sum(axis = 0)
    f_score = np.zeros((l, l))
    for i in range(l):
        for j in range(l):
            f_score[i, j] = (2*matriz_confusao[i, j])/(s_true[i]+s_predict[j])

    return ( (f_score.max(axis =1)*s_true).sum())/s
```
